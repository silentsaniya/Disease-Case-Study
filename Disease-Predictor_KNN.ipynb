```python
print("---Application of Saniya Shaikh---")

print("--=Diabetes Predictor Using KNN---")

#reading the file
df=pd.read_csv("/kaggle/input/disease-and-their-symptoms/dataset.csv")
df.head()
```

```python
#list of all unique diseases
dise=list(df["Disease"].unique())
dise
```

```python
#list of all unique symptoms
sym=df[["Symptom_1","Symptom_2","Symptom_3","Symptom_4","Symptom_5","Symptom_6","Symptom_7","Symptom_8","Symptom_9","Symptom_10","Symptom_11","Symptom_12","Symptom_13","Symptom_14","Symptom_15","Symptom_16","Symptom_17"]].stack().unique().tolist()
sym
```

```python
#filling NaN values:
for i in range(len(df)):
  y=df.iloc[i,1:18]                           
  v=y.mode()
  df.iloc[i,1:18].fillna(v[0],inplace=True)

df
```

```python
#importing relevant libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# splitting the dataset with 79% for training and 21% for testing
X=df.iloc[:,1:]
y=df.iloc[:,0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.21, random_state=42)
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
la=LabelEncoder()
X_train=X_train.apply(LabelEncoder().fit_transform)
X_test=X_test.apply(LabelEncoder().fit_transform)
```

```python
#fitting knn model 
from sklearn.neighbors import KNeighborsClassifier
neigh=KNeighborsClassifier(n_neighbors=32)
neigh.fit(X_train,y_train)
```

```python
#predicting
y_pred=neigh.predict(X_test)
```

```python
#function to replace unentered symptoms with NaN value 
def knnmodel(a=-1,b=-1,c=-1,d=-1,e=-1,f=-1,g=-1,h=-1,i=-1,j=-1,k=-1,l=-1,m=-1,n=-1,s=-1,w=-1,u=-1):
  v=neigh.predict([[a,b,c,d,e,f,g,h,i,j,k,l,m,n,s,w,u]])
  print(v)

knnmodel(2,21)
```

```python
#Performance Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')
```

```python
##printing the performance metrics
print("Accuracy: {:.2f}%".format(accuracy * 100)) #using {:.2f} to round off 2 decimal values
print("Precision: {:.2f}%".format(precision * 100))
print("Recall: {:.2f}%".format(recall * 100))
print("F1 Score: {:.2f}%".format(f1 * 100))
```
